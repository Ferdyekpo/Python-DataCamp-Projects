```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
```

```python
netflix_df.head(10)
```

```python
#Subsetting for just Movies in the dataset
netflix_subset = netflix_df[netflix_df['type'] == 'Movie']
```

```python
#Filtering for movies released after the 19's
subset = netflix_subset[(netflix_subset['release_year'] >= 1990)]
```

```python
#Streamlining the data for just movies in the 19's
movies_1990s = subset[(subset['release_year'] < 2000)]
movies_1990s.head()
```

```python
#Obtaining the most frequent count of movies.

plt.hist(movies_1990s['duration'])
plt.title('Distribution of movies Durattion in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of movies')

plt.show()
```

```python
#Number of short-action movies in the 1990's

subset_1 = movies_1990s[(movies_1990s['genre'] == 'Action')]

subset_1.head()
```

```python
#Iterating for movies with duration less than 90minutes of movies using a for loop

short_movie_count = 0

for label, row in subset_1.iterrows():
    if row['duration'] < 90 :
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count
        
print(short_movie_count)
```
